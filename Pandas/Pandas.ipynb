# merging data frames
it merge the comman attibutes in tables

```python
df1=pd.DataFrame({'s_names':['keshav','rajesh','prathamesh','Ashish'],'CGPA':[5,7,8,4]})
```

```python
df1
```

```python
df2=pd.DataFrame({'T_name':['nayan','pratik','keshav'],'CGPA':[2,4,5]})
```

```python
df2
```

```python
df1.merge(df2,how='inner')
```

```python
df1.merge(df2,'left')
```

```python
df1.merge(df2,how='right')
```

```python
df1.merge(df2,'outer') # combination of lefat join and left join
```
