```python
# import the require libraries
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris=sns.load_dataset('iris')
```

```python
iris
```

```python
iris.head()
```

```python
# to plot the petal_length
sns.displot(iris['petal_length'],bins=20,kde=True) # if you dont want line then make kde=False
plt.show()
```

```python
# if you want plot only line then write
sns.kdeplot(iris['petal_length'])
plt.show()
```

```python
# to make the scatter plot 
# if you want scatter petal length on x axis and petal width on y axis then 
sns.scatterplot(x=iris['petal_length'],y=iris['petal_width'])
plt.show()
```

# For coloring:
we use 'hue='parameter name''
<li>The 'hue' parameter in Seaborn histograms is a powerful tool for adding an extra layer of information to your plots. By using 'hue', you can easily compare the distributions of different categories within your data, making your visualizations more informative and insightful</li>

```python
# another way to write above code

sns.scatterplot(x='petal_length',y='petal_width',data=iris,hue='species')
plt.show()
```

# joinplot
<li>In Seaborn, sns.jointplot is used to create a multi-panel plot that combines univariate and bivariate visualizations. It is particularly useful for exploring the relationship between two variables, as it provides both their scatterplot (or other bivariate plot types) and marginal distributions.</li>

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris,kind='hex')
plt.show()
# if you want to and different kind then 
# kind=(hex,reg) etc
#  if you want to normal graph then remove kind parameter
```

```python
# if you want to plot all the combinaations  then you can 'pair plot'
sns.pairplot(data=iris,hue='species') # for colorining we use "hue=" parameter
plt.show()
```

# Categorical plots
Seaborn provides several categorical plot types that are specifically designed for visualizing categorical data. These plots are effective for exploring the relationships between categorical variables and numerical data.



```python
# it count the number of unique value in dataset
sns.countplot(x='species',data=iris,hue='species')
plt.show()
```

```python
# plt.figure(figsize=(7,7))  # to increse the size of the figure
sns.boxenplot(x='species',y='petal_length',data=iris,hue='species')
plt.show()
```

```python
sns.violinplot(x='species',y='petal_width',data=iris,hue='species')   # species is a catigirocal value
plt.show()
```
